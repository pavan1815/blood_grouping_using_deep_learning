```python
import numpy as np
import cv2
```

```python
x,y=[],[]
for i in images_dict['Agg']:
    img=cv2.imread(str(i))
    # print(img)
    resized=cv2.resize(img,(224,224))
    # print(resized)
    x.append(resized)
    y.append(1)# [im1,im2],[1,0]
for i in images_dict['nag']:
    img=cv2.imread(str(i))
    resized=cv2.resize(img,(224,224))
    x.append(resized)
    y.append(0)

x=np.array(x)
y=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=1)
x_train_scaled=x_train/255
y_train_scaled=y_train
x_test=x_test/255
print(x_train_scaled)
```

```python
os.mkdir("AUG_IMAGES_1")
```

```python
image_directory = '/content/ALL/'
SIZE = 224
dataset = []
my_images = os.listdir(image_directory)
for i, image_name in enumerate(my_images):    
    if (image_name.split('.')[1] == 'jpg'):        
        image = io.imread(image_directory + image_name)        
        image = Image.fromarray(image, 'RGB')        
        image = image.resize((SIZE,SIZE))
        dataset.append(np.array(image))
x = np.array(dataset)
i = 0
for batch in datagen.flow(x, batch_size=16,
                          save_to_dir= '/content/AUG_IMAGES_1',
                          save_prefix='dr',
                          save_format='jpg'):    
    i += 1
    if i > 50:        
        break
```

```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras.applications.vgg16 import VGG16
```

```python
base_model = VGG16(input_shape = (224, 224, 3),
include_top = False,
weights = 'imagenet')

for layer in base_model.layers:
    layer.trainable = False
x = layers.Flatten()(base_model.output)
x = layers.Dense(512, activation='relu')(x)
x = layers.Dropout(0.5)(x)
x = layers.Dense(2, activation='sigmoid')(x)
modelv16 = tf.keras.models.Model(base_model.input, x)
modelv16.compile(optimizer = 'adam', loss = 'categorical_crossentropy',metrics = ['acc'])
modelv16.summary()
```

```python
modelv16.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
epochs=7
history=modelv16.fit(x=datagen.flow(x_train_scaled,y_train_scaled),batch_size=48, epochs=10,steps_per_epoch=3)
```

```python
# FAST_RUN = False
IMAGE_WIDTH=128
IMAGE_HEIGHT=128
IMAGE_SIZE=(IMAGE_WIDTH, IMAGE_HEIGHT)
IMAGE_CHANNELS=3
```

```python
import keras,os
from keras.models import Sequential
from keras.layers import Dense, Conv2D, MaxPool2D , Flatten
from keras.preprocessing.image import ImageDataGenerator
import numpy as np
from keras.models import Sequential
from keras.layers import Conv2D,MaxPooling2D,Convolution2D,Activation,Flatten,Dense,Dropout,MaxPool2D,BatchNormalization
# from keras.utils import to_categorical
from keras.preprocessing.image import ImageDataGenerator
```

```python
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```python
earlystop = EarlyStopping(patience=10)
```

```python
learning_rate_reduction= ReduceLROnPlateau(monitor='val_accuracy', 
                                            patience=2, 
                                            verbose=1, 
                                            factor=0.3, 
                                            min_lr=0.00001)
```

```python
callbacks = [earlystop,learning_rate_reduction]
```

```python
df["category"] = df["category"].replace({0: 'Non_agglutination', 1: 'Agglutination'})
```

```python
# splitting data for training and validation
df = df.sample(frac = 1)
train_df, validate_df = train_test_split(df, test_size=0.25, random_state=42)
train_df = train_df.reset_index(drop=True)
validate_df = validate_df.reset_index(drop=True)
```

```python
train_df['category'].value_counts().plot.bar()
```

```python
validate_df['category'].value_counts().plot.bar()
```

```python
#count of traning and validation
total_train = train_df.shape[0]
total_validate = validate_df.shape[0]
# print(total_validate)
batch_size=32
```

```python
train_datagen = ImageDataGenerator(
    rotation_range=45,
    shear_range=0.1,
    rescale=1./255,
    zoom_range=0.2,
    brightness_range=(0.5,0.8),
    horizontal_flip=True,
    vertical_flip=True,
    width_shift_range=0.1,
    height_shift_range=0.1,
    fill_mode='nearest',
    # interpolation_order=1
)

train_generator = train_datagen.flow_from_dataframe(
    train_df, 
    "/content/ALL", 
    x_col='filename',
    y_col='category',
    target_size=IMAGE_SIZE,
    class_mode='categorical',
    batch_size=batch_size
)
```

```python
len(os.listdir("/content/ALL"))
```

```python
validation_datagen = ImageDataGenerator(rescale=1./255)
validation_generator = validation_datagen.flow_from_dataframe(
    validate_df, 
    "/content/ALL", 
    x_col='filename',
    y_col='category',
    target_size=IMAGE_SIZE,
    class_mode='categorical',
    batch_size=batch_size
)
```

```python
example_df = train_df.sample(n=1).reset_index(drop=True)
example_generator = train_datagen.flow_from_dataframe(
    example_df, 
    "/content/ALL", 
    x_col='filename',
    y_col='category',
    target_size=IMAGE_SIZE,
    class_mode='categorical'
)
```

```python
plt.figure(figsize=(12, 12))
for i in range(0, 15):
  plt.subplot(5, 3, i+1)
  for X_batch,Y_batch in example_generator:
    image = X_batch[0]
    plt.imshow(image)
    break
plt.tight_layout()
plt.show()
```

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, Activation, BatchNormalization

model = Sequential()

model.add(Conv2D(32, (3, 3), activation='relu', input_shape=(IMAGE_WIDTH, IMAGE_HEIGHT, IMAGE_CHANNELS)))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

# model.add(Conv2D(64, (3, 3), activation='relu'))
# model.add(BatchNormalization())
# model.add(MaxPooling2D(pool_size=(2, 2)))
# model.add(Dropout(0.25))

model.add(Conv2D(128, (3, 3), activation='relu'))
model.add(BatchNormalization())
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Flatten())
model.add(Dense(512, activation='relu'))
model.add(BatchNormalization())
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax')) # 2 because we have cat and dog classes

model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])

model.summary()
    
```

```python
batch_size=48
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
epochs= 5
model.summary()
```

```python
history = model.fit(
    train_generator, epochs=20,
    validation_data=validation_generator,
    # validation_steps=total_validate//batch_size,
    steps_per_epoch=total_train//batch_size,
    callbacks=callbacks
)
```

```python
model.evaluate(validation_generator)
```
